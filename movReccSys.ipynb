```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re  #regular repression
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    text = str(text).lower()  #lowercase
    text = re.sub(r'[^a-zA-Z\s]', '', text)  #punctuation removal (keep only a-z and A-Z)
    
    words = text.split() # split text based on space
    words = [word for word in words if word not in stop_words]  #remove stopwords

    words = [lemmatizer.lemmatize(word) for word in words]  #lemmatize

    return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
# Text Vectorization
```

```python
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000, ngram_range=(1,2), stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
# cosine similarity
```

```python
# Cosine similarity is a way to measure how similar two texts
# (or vectors) are by comparing their angle, not their length.

# 🧠 Intuition
# Think of each document as a vector (like from TF-IDF)
# Cosine similarity checks how aligned they are
# Value ranges from:
# 1 → exactly same direction (very similar)
# 0 → no similarity (orthogonal)
# -1 → opposite (rare in text)

# 📐 Formula
# cosine similarity=
# A⋅B / ∣∣A∣∣×∣∣B∣∣

# A⋅B → dot product
# ∣∣A∣∣,∣∣B∣∣ → magnitudes
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n=10):
    if title not in indices:
        return ['Movie not found']
    idx = indices[title] 
    sim_scores = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
    similar_idx = sim_scores.argsort()[::-1][1:n+1]
    return df['title'].iloc[similar_idx]
```

```python
recommend('Toy Story')
```

```python
recommend('Avengers')
```

```python
recommend('Avenger')
```

```python
recommend('Avatar', 5)
```
